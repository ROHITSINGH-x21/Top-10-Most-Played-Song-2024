```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Here is Data of Most Popular/Played Song in 2024, We need to Show top 10 Most Played Songs.
#1 Firstly we import the data

```python
df = pd.read_csv('Spotify.csv')
```

#2 Then we need to clean data, for that we need to drop unnessecery columns 

```python
df.drop(['ISRC', 'Track Score', 
       'Spotify Playlist Count', 'Spotify Playlist Reach',
       'YouTube Likes', 'TikTok Posts',
       'TikTok Likes', 'YouTube Playlist Reach',
       'Apple Music Playlist Count', 'AirPlay Spins', 'SiriusXM Spins',
       'Deezer Playlist Count', 'Deezer Playlist Reach',
       'Amazon Playlist Count', 'Pandora Streams', 'Pandora Track Stations',
       'Soundcloud Streams', 'Shazam Counts', 'TIDAL Popularity','Spotify Popularity','Explicit Track'],axis='columns',inplace=True)
```

```python
df.head()
```

#3 after some cleaning we need to know Shape, Info, Describe and Null Values 

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.describe().T
```

```python
df.info()
```

#4 Here we removing commas from the values in specified columns

```python
views_list = ['Spotify Streams'	,'YouTube Views','TikTok Views']
for x in views_list:
    df[x]=df[x].str.replace(',','')
```

#5 We got the info from 'isnull' that in these column their are so much null value, So we need to fill them with '0'

```python
df.fillna(0,inplace=True)
```

```python
df.isnull().sum()
```

#6 After Geting Info we got that ('Spotify Streams'	,'YouTube Views','TikTok Views') these 3 columns are object.
for counting Views/streams we need data in 'int' form.
so we need chnage their data type for (object) to (int).

```python
for x in views_list:
    df[x] = df[x].astype(np.int64)
```

#7 after changing data type we create a new column for total views using of these 3 Columns

```python
df['Total Views'] = df['Spotify Streams'] +	 df['YouTube Views'] + df['TikTok Views']
```

```python
df
```

#8 then We need to ascend data from high to low to get top 10 Songs

```python
final_data = df.sort_values(by='Total Views',ascending=False).head(10)
```

```python
final_data
```

#9 Here is the visuals of the top 10 most played song in 2024

```python
labels = final_data.Track
plt.style.use('ggplot')
colors = ['#e6194b', '#3cb44b', '#ffe119', '#4363d8', '#f58231', '#911eb4', '#46f0f0', '#f032e6', '#bcf60c', '#fabebe']
fig=plt.pie(final_data['Total Views'],pctdistance=01.1,autopct='%.1f %%',colors=colors,)
plt.legend(fig[0], labels, loc="upper left", bbox_to_anchor = (1.4,1))
plt.title('Top 10 Most Streamed Songs in 2024')
plt.show()
```
